```python
from bs4 import BeautifulSoup
import requests 
```

```python
#accessing url and geting data 
html_text  = requests.get("https://www.worldometers.info/coronavirus").text
code = BeautifulSoup(html_text, 'lxml')
```

```python
#finding tables
code =code.table 
```

```python
# finding tr tags from html 
tags = code.find_all('tr')
# for tag in tags:
#     x=tag.text
#     print(x)
```

```python
data = []
for tag in tags:
    x = tag.text
    data.append(x)
```

```python
updated_data = []
for i in data:
    updated_data.append(i.split('\n')[1:-1])
```

```python
cleaned_data = [ i for i in updated_data if i[0]!=''] # removing empty 
```

```python
cleaned_data
```

```python
import csv

with open('new_data.csv','w')as f:
    x = csv.writer(f)
    for i in cleaned_data:
        x.writerow(i)

    
```
